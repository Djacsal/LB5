#### 2.8 работа с пропущенными значениями (если они есть): удаление строк с пропущенными значениями, заполнение пропущенных значений средним значением по колонке. Если пропущенных значений нет — намеренно их "генерируете", прибивая какие-то куски данных в np.nan

```python
# Проверка пропущенных значений:
table.isnull().sum()
```

```python
# Удаление строк с пропущенными значениями:
table = table.dropna(axis=0)
print("После удаления:")
table
```

```python
# Намеренно сгенерируем пропущенные значения, прибивая какие-то куски данных в np.nan:
import numpy as np
table.iloc[1:10, 6:7] = np.nan
table
```

```python
# Заполнение пропущенных значений средним значением по колонке:
numeric_cols = [ 'Возраст']

# заполняем пропущенные значения средним по столбцу:
table[numeric_cols] = table[numeric_cols].fillna(table[numeric_cols].mean())
print("Полсе заполнения:")
table
```

#### 2.9. создание нового поля вычисленного на основе значений других полей: 
#### 2.9.1 через выражение на базе имеющихся колонок


```python
# создаем новый столбец: "Начало работы"
table['Начало работы'] = table['Возраст'] - table['Стаж работы']
table
```

#### 2.9.2 через DataFrame.apply

```python
# Создаём новый столбец: "Стаж в месяцах", если стаж равен нулю, то мы возвращаем NaN.
table['Стаж в месяцах'] = table.apply(lambda row: row['Стаж работы'] * 15 if row['Стаж работы'] > 0 else np.nan, axis=1)
table
```

#### 2.9.3 через Series.apply


```python
# Создаём новый столбец: "Сортировка по возрасту", которое определяется на основе значений столбца "Возраст". 
table['Сортировка по возрасту'] = table['Стаж работы'].apply(lambda x: 'Модератор' if x <= 15  else 'Эксперт' if x > 15 else 'Неопределено')
table
```

#### 2.10 сортировка по какому-либо из полей

```python
# Сортировка по полю "Возраст" в порядке возрастания
sorted_table = table.sort_values(by='Возраст', ascending=True)
sorted_table
```

```python
# Сортировка по полю "Стаж работы" в порядке убывания
sorted_table2 = table.sort_values(by='Стаж работы', ascending=False)
sorted_table2
```
